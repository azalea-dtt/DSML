```python
df = pd.read_csv('/kaggle/input/horse-colic/horse.csv')
df.info()
target = df['outcome']
df = df.drop(['outcome'], axis=1)
# drop column
# inplace == False --> return a copy
```

```python
#df.columns.tolist()
df = pd.get_dummies(df, prefix_sep='_', drop_first=True,dummy_na=False)
```

```python
# Read: https://towardsdatascience.com/what-and-why-behind-fit-transform-vs-transform-in-scikit-learn-78f915cf96fe
#fit_transform: for training data, calculate the mean, std, scaling the training data and let the model learn these variables as well
# transform: scaling the test data
# if we use fit_transform for the test data, the model will also learn the mean, std of test data
# but we want the test data was not trained before, so the model does not know about test data, thus if the prediction accuracy on the test data is good, it means that it's reliable
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
#Encode target labels with value between 0 and n_classes-1.
#This transformer should be used to encode target values, i.e. y, and not the input X.
```

```python
X,y = df.values, target.values
```

```python
label_encoder = LabelEncoder()
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.33, random_state=42)
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
X_train = imp.fit_transform(X_train)
X_test = imp.transform(X_test)
df.isnull().sum()
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

from sklearn.metrics import accuracy_score,classification_report 
print("Decision Tree Accuracy: ",accuracy_score(y_test, y_pred)*100)
```
